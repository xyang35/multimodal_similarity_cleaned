```python
import pickle
import matplotlib.pyplot as plt
import glob
import os
import numpy as np

import sys
sys.path.append('../')
from preprocess.honda_labels import honda_labels2num, honda_num2labels

label_dir = '/mnt/work/honda_100h/labels/'

for key, value in honda_num2labels.items():
    print ("%d: %s" % (key, value))
```

```python
# Event instance distribution

files = glob.glob(label_dir+'2017*')
counts = np.zeros((len(files), 13))

for i in range(len(files)):
    label = pickle.load(open(files[i], 'rb'))
    for l in label['G']:
        counts[i, l] += 1
print (counts)

counts_sum = np.sum(counts, axis=0)
counts_avg = np.mean(counts, axis=0)
counts_std = np.std(counts, axis=0)
for i in range(13):
    print ("%d: sum=%d, avg=%f (+-%f)" % (i, counts_sum[i], counts_avg[i], counts_std[i]))
print ("Total: %d" % np.sum(counts_sum))

events_per_sess = np.sum(counts, axis=1)
for i in range(events_per_sess.shape[0]):
    print (files[i], events_per_sess[i])
```
